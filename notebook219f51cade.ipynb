```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import sklearn as sk
from sklearn.linear_model import LinearRegression
```

```python
data = pd.read_csv("../input/unflood2/unflood - Sheet1.csv")

X = data.drop(['Rainrisk','Water level'], axis=1)
Y = data['Water level']
```

```python
X[:10]
```

```python
X_train = X.drop([X.index[3], X.index[4]], axis=0)
X_train
```

```python
X_train[:8].values
```

```python
Y_train = Y.drop([Y.index[3], Y.index[4]], axis=0)
Y_train
```

```python
Y_train = Y_train.values.reshape(-1,1)
```

```python
Y_train[0:8]
```

```python
lr = LinearRegression()
lr.fit(X_train[:8].values,Y_train[:8])
```

```python
inp = np.array([[36],[26],[6],[0],[72]])
inp = inp.reshape(1, -1)
print(inp)
```

```python
print("Predicted Water Lvl: ",lr.predict(inp))
```

```python
lr.score(X_train[:8].values,Y_train[:8])
```

```python
import pickle
filename = 'finalized_model.sav'
pickle.dump(lr, open(filename, 'wb'))
```
